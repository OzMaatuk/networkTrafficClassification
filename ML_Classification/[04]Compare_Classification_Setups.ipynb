# Compare Classification Setups For All Algorithms
presenting the effect of diffrent data setups for all the available algorithms

# Import

```python
import pandas as pd
from pycaret.classification import *
```

# Read Data
using all of the data to get teh sense of what learning algorithm should we use

```python
# set target feature
target_label = 'tuple'
# set learning models
#learning_model = ['rf'] # ['rf','lightgbm','et','dt','knn','xgboost']
# set numeric features which pycaret takes as category
num_features = ['min_packet_size', 'min_fpkt', 'min_bpkt']
path = "../Datasets/" + target_label + "_dataset/"
```

```python
data = pd.read_csv(path+r'all_features_'+target_label+'.csv',
                      sep='\t',
                      skiprows=[1])
```

# Clear Setup And Compare

```python
setup(data = data,
        target=target_label,
        silent=True,
        numeric_features=num_features)
compare_models()
```

# Change Imputation And Compare

```python
setup(data = data,
        target=target_label,
        silent=True,
        numeric_features=num_features,
        numeric_imputation='median')
compare_models()
```

# Base Normlized Setup And Compare

```python
setup(data = data,
        target=target_label,
        silent=True,
        numeric_features=num_features,
        normalize=True)
compare_models()
```

# Advance Normlized Setup And Compare

```python
setup(data = data,
        target=target_label,
        silent=True,
        numeric_features=num_features,
        normalize=True,
        transformation=True,
        transformation_method='quantile')
compare_models()
```
