# Compare Learning Algorithms
This notebook compare all the available learning algorithms in the task of classifying (predicting) the "tuple" < Browser, Operation-System, Application > (BOA) and present prediction scores.

# Import

```python
import pandas as pd
from pycaret.classification import *
```

# Read Data
using all of the data to get teh sense of what learning algorithm should we use

```python
# set target feature
target_label = 'tuple'
num_features = ['min_packet_size', 'min_fpkt', 'min_bpkt']
path = "../Datasets/" + target_label + "_dataset/"
data_file = "new_all_features_"
```

```python
data = pd.read_csv(path+data_file+target_label+'.csv',
                      sep='\t',
                      skiprows=[1])
```

# Setup And Compare

```python
setup(data = data,
        target=target_label,
        silent=True,
        numeric_features=num_features)
compare_models()
```

finally we can note that eXtream-Gradient-Boosting-Machine is the most accurate algorithm, but when cosidering also the prediction time, Extra-Trees if the prefered one.
