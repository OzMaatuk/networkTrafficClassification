# Import and Constants

```python
import pandas as pd
import numpy as np
from pycaret.classification import *
import time
```

```python
# set constants
target_label = 'tuple'
learning_model = 'et'
num_features = ['min_packet_size', 'min_fpkt', 'min_bpkt']
file_name = "new_all_features_"
path = "../Datasets/" + target_label + "_dataset/"
```

# Read Data

```python
data = pd.read_csv(path+r'new_all_features_'+target_label+'.csv',
                      sep='\t',
                      skiprows=[1])
```

```python
data.shape
```

```python
data[target_label].unique()
```

# Setup Data and Create Model

```python
setup(data=data,
      target=target_label,
      numeric_features=num_features,
      silent=True)
model=create_model(learning_model)
```
