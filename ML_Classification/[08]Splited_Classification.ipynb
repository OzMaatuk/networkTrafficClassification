# Split Tuple
this notebook testing the difficulty of classifying the three labels of Browser, Operation-System, Application (BOA) verses classifying the labels separately.

# Import

```python
import pandas as pd
from pycaret.classification import *
import time
```

# Settings

```python
# set constants
target_label = ''
learning_model = 'et' #['rf','et','lightgbm','xgboost']
num_features = ['min_packet_size', 'min_fpkt', 'min_bpkt']
file_name = "converted_all_features_"
path = "../Datasets/" + target_label + "_dataset/"
os_prediction = []
browser_prediction = []
app_prediction = []
```

```python
# function for making model-prediction over the data set and measure the run time 
def timed_prediction(in_data,in_model):
    t = time.process_time()
    predicted = predict_model(in_model, data=in_data)
    elapsed_time = time.process_time() - t
    print("prediction took: " + str(elapsed_time))
    return predicted
```

```python
# function for checkign the correction of the model-prediction over the data
def check_correction(in_predicted):
    count=0
    index = in_predicted.index
    number_of_rows = len(index)
    for i in range(0,number_of_rows):
        if str(int(in_predicted.iloc[i][target_label])) != str(int(in_predicted.iloc[i]['Label'])):
            #print("prediction not matched in line " + str(i) + " as " + str(in_predicted.iloc[i]['app']) + "!=" + str(in_predicted.iloc[i]['Label']))
            count=count+1
    print("number of error: " + str(count) + " from " + str(number_of_rows) + " test samples \n which is " + str(100*count/number_of_rows) + " precent of error.")
```

# OS
read proper data, setup Classifier, Compare Models and Predicte.

```python
target_label = 'os'
path = "../Datasets/" + target_label + "_dataset/"
```

```python
os_data = pd.read_csv(path+file_name+target_label+'.csv',
                      sep='\t',
                      skiprows=[1])
```

```python
setup(data=os_data,
      target=target_label,
      silent=True)
model=create_model(learning_model)
```

```python
predicted = timed_prediction(os_data,model)
```

```python
os_prediction = predicted['Label']
```

# Compare prediction with original labels

```python
col_len = len(os_data['os'])
os_count = 0
browser_count = 0
app_count = 0
total_count = 0
for i in range(0,col_len):
    if int(os_prediction[i]) != int(os_data['os'][i]):
        total_count=total_count+1
        os_count=os_count+1
        continue
    if int(browser_prediction[i]) != int(browser_data['browser'][i]):
        total_count=total_count+1
        browser_count=browser_count+1
        continue
    if int(app_prediction[i]) != int(app_data['app'][i]):
        total_count=total_count+1
        app_count=app_count+1
        continue
print("total error count = " + str(total_count))
print("it is " + str(100*total_count/col_len) + "% of the data")
print("error in os = " + str(os_count))
print("error in browser = " + str(browser_count))
print("error in app = " + str(app_count))
```
